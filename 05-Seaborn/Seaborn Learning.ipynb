# Seaborn Learning

Seaborn is a popular Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive, informative, and complex statistical graphics with minimal code. It is highly integrated with Pandas data structures and makes data exploration much easier.

```python
import  seaborn as  sns
```

```python
Tips=sns.load_dataset('tips')
Tips
```

```python
import matplotlib.pyplot as plt
```

# Scatter Plot

```python
sns.scatterplot(x= 'total_bill' , y= 'tip' , data= Tips)
plt.title('Scatter Plot of Tip vs Total Bill')
```

```python
plt.figure(figsize=(8,6))

sns.regplot(
    x='total_bill',
    y='tip',
    data=Tips,
    scatter_kws={'alpha':0.6, 's':70},
    line_kws={'color':'red'}
)

plt.title('Relationship Between Total Bill and Tip', fontsize=15)
plt.grid(True)
plt.tight_layout()
```

# Line Plot

```python
sns.lineplot(x='size',y='total_bill',data=Tips)
plt.title('Line Plot of Size vs total Bill')
```

```python
plt.figure(figsize=(10,6))

sns.lineplot(
    x='size',
    y='total_bill',
    data=Tips,
    marker='o',
    linewidth=3,
    markersize=8,
    color='royalblue'
)

plt.title('Average Total Bill by Group Size', fontsize=16, fontweight='bold')
plt.xlabel('Group Size', fontsize=12)
plt.ylabel('Total Bill ($)', fontsize=12)

plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

# Bar Plot

```python
sns.barplot(x='day',y='total_bill',data=Tips)
plt.title('Bar Plot of Days vs Total Bill')
```

```python
plt.figure(figsize=(9,6))

ax = sns.barplot(
    x='day',
    y='total_bill',
    data=Tips,
    palette='Set2',
    edgecolor='black'
)

for container in ax.containers:
    ax.bar_label(container, fmt='%.1f')

plt.title('Average Total Bill by Day', fontsize=16, fontweight='bold')
plt.xlabel('Day')
plt.ylabel('Average Total Bill ($)')
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

# Box Plot

```python
sns.boxplot(x='day',y='total_bill',data=Tips)
plt.title('Box Plot of Days vs Total Bill')
```

```python
plt.figure(figsize=(10,6))

sns.set_style("whitegrid")

sns.boxplot(
    x='day',
    y='total_bill',
    data=Tips,
    palette='Set2',
    linewidth=2,
    width=0.6
)

plt.title('Distribution of Total Bill by Day', fontsize=16, fontweight='bold')
plt.xlabel('Day', fontsize=12)
plt.ylabel('Total Bill ($)', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

# Violin Plot

```python
sns.violinplot(x='day',y='total_bill',data=Tips)
plt.title('Violin Plot of Days vs Total Bill')
```

```python
plt.figure(figsize=(10,6))

sns.set_style("whitegrid")

sns.violinplot(
    x='day',
    y='total_bill',
    data=Tips,
    palette='Set2',
    linewidth=2
)

plt.title('Distribution of Total Bill by Day', fontsize=16, fontweight='bold')
plt.xlabel('Day', fontsize=12)
plt.ylabel('Total Bill ($)', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

# Histogram Plot

```python
 sns.histplot(Tips['total_bill'],bins=10,kde=True)
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    Tips['total_bill'],
    bins=10,
    kde=True,
    color='royalblue',
    edgecolor='black',
    linewidth=1.2
)

plt.title('Distribution of Total Bill', fontsize=16, fontweight='bold')
plt.xlabel('Total Bill ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

# KDE Plot

```python
sns.kdeplot(Tips['total_bill'],shade=True)
```

```python
plt.figure(figsize=(10,6))

sns.set_style("whitegrid")

sns.kdeplot(
    data=Tips,
    x='total_bill',
    fill=True,
    color='royalblue',
    linewidth=2
)

plt.title('Density Distribution of Total Bill', fontsize=16, fontweight='bold')
plt.xlabel('Total Bill ($)', fontsize=12)
plt.ylabel('Density', fontsize=12)

plt.grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

# Pair Plot

```python
sns.pairplot(Tips)
```

# Coorelation

```python
corr=Tips[['total_bill','tip','size']].corr()
corr
```

# HeatMap

```python
sns.heatmap(corr,annot=True)
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    fmt='.2f',
    linewidths=1,
    linecolor='black'
)

plt.title('Correlation Heatmap', fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```

# Bike Data Sales Visulalization

```python
import pandas as pd
BS_df=pd.read_csv('Bike_Sales_Data.csv')
```

```python
BS_df.head(5)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Country',y='Profit',data=BS_df,estimator=sum)
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    x='Country',
    y='Profit',
    data=BS_df,
    estimator=sum,
    palette='Set2'
)

plt.title('Total Profit by Country', fontsize=15)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Profit', fontsize=12)

plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=45)

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
